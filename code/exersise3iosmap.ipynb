```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import pandas as pd

from sklearn.manifold import MDS,Isomap, LocallyLinearEmbedding
```

```python
def Lorenz_function(l,t,a,b,c):
    l1,l2,l3 = l.tolist()
    return a*(l2-l1),l1*(c-l3)-l2,l1*l2-b*l3

a = 10
b = 8/3
c = 20
t = np.arange(-100,1000,0.01)
# t = np.arange(-100,1000,0.1)
y0 = np.array([1,2,3])
y1 = integrate.odeint(Lorenz_function,y0,t,args=(a,b,c))
y2 = y1[10000:]

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(projection='3d')
ax.plot(y1[:,0],y1[:,1],y1[:,2],color='b')
# ax.plot(y2[:,0],y2[:,1],y2[:,2],color='b')
plt.xlabel('a1')
plt.ylabel('a2')
ax.set_zlabel('a3')
plt.show()
```

```python
df = pd.DataFrame(y1)
df.to_excel('data.xlsx',index=False)
```

```python
isomap = Isomap(n_components=2, n_neighbors=10)

X_isomap = isomap.fit_transform(y2)
```
